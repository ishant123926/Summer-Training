```python
import pandas as pd
import numpy as np
data = {
    "Date": pd.date_range(start="2020-03-01", periods=5).tolist() * 3,
    "Country": ["USA"]*5 + ["India"]*5 + ["Brazil"]*5,
    "Confirmed": [1000, 1500, 2200, 3000, 4100, 500, 800, 1200, 1900, 2500, 300, 600, 900, 1400, 2000],
    "Recovered": [200, 400, 700, 1100, 1600, 100, 250, 500, 800, 1200, 50, 150, 300, 550, 900],
    "Deaths": [20, 35, 50, 75, 110, 10, 15, 25, 40, 60, 12, 22, 35, 55, 80]
}

df = pd.DataFrame(data)
```

```python
import pandas as pd
df['Recovery_Rate (%)'] = (df['Recovered'] / df['Confirmed']) * 100
df['Death_Rate (%)'] = (df['Deaths'] / df['Confirmed']) * 100

df = df.sort_values(by="Date")
country_comparison = df.groupby("Country").agg({
    "Confirmed": "last",
    "Recovered": "last",
    "Deaths": "last",
    "Recovery_Rate (%)": "last",
    "Death_Rate (%)": "last"
}).reset_index()
country_comparison = country_comparison.sort_values(by="Confirmed", ascending=False)

print("Country-wise Summary (Latest Data):")
print(country_comparison)
```

```python
high_fatality = df[df["Death_Rate (%)"] > 3.0]
```

```python
df = df.sort_values(by=["Country", "Date"])


df["New_Daily_Cases"] = df.groupby("Country")["Confirmed"].diff().fillna(df["Confirmed"])
```

```python
country_comparison = df.groupby("Country").agg({
    "Confirmed": "max",
    "Recovered": "max",
    "Deaths": "max",
    "Recovery_Rate (%)": "last",
    "Death_Rate (%)": "last"
}).reset_index()

print("\nCountry-wise Summary:")
print(country_comparison)
```
